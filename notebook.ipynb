### Standardization and Scaling

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_scaled.mean(axis=0) 
X_scaled.std(axis=0) # Mean of each feature after scaling
```

## Random Forest Classifier
Let's use a Random Forest Classifier to predict the `price_range` label.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

# Split the data for validation
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize and train the Random Forest Classifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Predict on validation set
y_pred = rf.predict(X_test)

# Evaluate the model
acc = accuracy_score(y_test, y_pred)
print(f"Validation Accuracy: {acc:.4f}")
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(x=rf.feature_importances_, y=X.columns)
plt.title('Feature Importances from Random Forest')
plt.show()
```

## Logistic Regression for Multiclass Classification
Let's use logistic regression to predict the `price_range` label as a multiclass problem.

```python
from sklearn.linear_model import LogisticRegression

# Initialize and train the Logistic Regression model for multiclass
logreg = LogisticRegression(solver='lbfgs', max_iter=1000, random_state=42)
logreg.fit(X_train, y_train)

# Predict on validation set
y_pred_logreg = logreg.predict(X_test)

# Evaluate the model
acc_logreg = accuracy_score(y_test, y_pred_logreg)
print(f"Validation Accuracy: {acc_logreg:.4f}")
print(classification_report(y_test, y_pred_logreg))
```

```python
fig, axes = plt.subplots(4, 1, figsize=(12, 40))
axes = axes.flatten()
class_labels = logreg.classes_

for i, ax in enumerate(axes):
    sns.barplot(x=logreg.coef_[i], y=X.columns, ax=ax)
    ax.set_title(f'Logistic Regression Coefficients: Class {class_labels[i]}')
    ax.set_xlabel('Coefficient Value')
    ax.set_ylabel('Feature')

plt.tight_layout()
plt.show()
```

## XGBoost Classifier
Let's use XGBoost to predict the `price_range` label and compare its performance.

```python
from xgboost import XGBClassifier

# Initialize and train the XGBoost Classifier
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
xgb.fit(X_train, y_train)

# Predict on validation set
y_pred_xgb = xgb.predict(X_test)

# Evaluate the model
acc_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"Validation Accuracy: {acc_xgb:.4f}")
print(classification_report(y_test, y_pred_xgb))
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(x=xgb.feature_importances_, y=X.columns)
plt.title('Feature Importances from XGBoost')
plt.show()
```

## Support Vector Machine (SVM) Classifier
Let's use an SVM to predict the `price_range` label and compare its performance.

```python
from sklearn.svm import SVC

# Initialize and train the SVM Classifier
svm = SVC(kernel='rbf', random_state=42)
svm.fit(X_train, y_train)

# Predict on validation set
y_pred_svm = svm.predict(X_test)

# Evaluate the model
acc_svm = accuracy_score(y_test, y_pred_svm)
print(f"Validation Accuracy: {acc_svm:.4f}")
print(classification_report(y_test, y_pred_svm))
```
